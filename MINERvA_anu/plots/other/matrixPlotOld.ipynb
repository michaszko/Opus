```python
"""ploting mine or Tomek's matrix with values on it - MINERvA_anu"""

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm

logTrue=0
textTrue=1
tomekTrue=0

data=np.array([])

if tomekTrue:
    title='tomek'
else:
    title='chi2cov_154.272_maxIter_1_continuous_0.4_InteriorPoint'

with open('../../results/chi2Values/'+title+'.dat') as file:
    data = [[float(digit) for digit in line.split()] for line in file]

fig = plt.figure(num=1, figsize=(1, 1), dpi=1, facecolor='w', edgecolor='k')
    
fig, ax = plt.subplots()

if textTrue:
    for (j,i),label in np.ndenumerate(data):
        ax.text(50*i+25,50*j+25,"{0:.1f}".format(label),ha='center',va='center',fontsize=3,color="black",weight='bold')

if logTrue:
    im = ax.imshow(data, origin="lower",extent=(1,1200,1,1200),norm=LogNorm(vmin=0.01, vmax=60))
    
else:
    im = ax.imshow(data, origin="lower",extent=(1,1200,1,1200),vmax=2)

cbar = ax.figure.colorbar(im, ax=ax)
cbar.ax.set_ylabel(" ", rotation=-90, va="bottom")

ax.set_ylabel('$\omega$ [MeV]')
ax.set_xlabel('$q$ [MeV]')
ax.set_title(title+"\n",fontsize=10)



if logTrue:
    fig.savefig('MINERvA_nu_' + title + '_log.pdf', dpi=300, bbox_inches='tight', pad_inches=0, quality=95)
    
else:
    fig.savefig('MINERvA_nu_' + title + '.pdf', dpi=300, bbox_inches='tight', pad_inches=0, quality=95)

plt.show()
```
